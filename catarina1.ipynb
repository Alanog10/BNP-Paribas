### Checking outliers

```python
def plot_multiple_boxplots(data, feats, n_cols=3, title="Numeric Variables' Box Plots (>0 only)"):
    
    n_rows = ceil(len(feats) / n_cols)
    
    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(6 * n_cols, 4.5 * n_rows)
    )
    
    axes = axes.flatten()
    
    for i, feat in enumerate(feats):
        
        # Keep only values strictly greater than 0
        values = data[feat]
        values = values[(values > 0) & (values.notna())]
        
        if len(values) > 0:
            sns.boxplot(x=values, ax=axes[i])
            axes[i].set_title(feat, fontsize=10)
        else:
            axes[i].set_title(f"{feat} (no >0 values)", fontsize=10)
    
    # Remove unused subplots
    for j in range(len(feats), len(axes)):
        fig.delaxes(axes[j])
    
    plt.suptitle(title, fontsize=18)
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
```

```python
plot_multiple_boxplots(FAMA, num_cols)
```

### Feature Engineering

```python
# proporção de créditos ativos

FAMA["active_credit_ratio"] = (
    FAMA["ALLBD_A_CL__N"] + FAMA["ALLBD_A_CP__N"]
) / (
    FAMA["ALLBD_N_CL__N"] + FAMA["ALLBD_N_CP__N"] + 1
)
```

0 -> nenhum credito ativo

1 -> todos os créditos estão ativos

```python
# carga financeira (montante de crédito vs rendimento)
FAMA["debt_income_ratio"] = (
    FAMA["ALLBD_MTFIN_CL__N"] /
    (FAMA["ALLBD_mean_active_resso__N"] + 1)
)
```

```python
# ratio de progresso do crédito

FAMA["credit_progress"] = (
    FAMA["ALLBD_MTFIN_CL__N "] /
    (FAMA["ALLBD_MTFINO_CL__N "] + 1)
)
```

```python
# percentagem de credito por pagar

FAMA["credit_remaining_ratio"] = (
    FAMA["ALLBD_MTFIN_CL__N"] /
    (FAMA["ALLBD_MTFINO_CL__N"] + 1)
)
```

```python

BDOSS = BDOSS.copy()

# Converter datas
BDOSS['OBS_DATE'] = pd.to_datetime(BDOSS['OBS_DATE'])

# Normalizar POS
BDOSS['POS'] = BDOSS['POS'].str.strip().str.upper()

# Ordenar painel
BDOSS = BDOSS.sort_values(['DOSSIER', 'OBS_DATE'])
```

```python
df_last_obs = (
    BDOSS
    .groupby('DOSSIER', as_index=False)
    .tail(1)
    .copy()
)

print("Total contracts:", BDOSS['DOSSIER'].nunique())
print("Lines pos tail(1):", len(df_last_obs))

# para agrupar por contrato e identificar a ultima linha de cada
```

```python
df_term = df_last_obs[df_last_obs['POS'].isin(['SAN', 'SOL'])].copy()

print(df_term['POS'].value_counts())

# filtrar por contratos com POS SAN ou SOL
```

```python
df_term['early_settlement'] = (df_term['POS'] == 'SAN').astype(int)

df_term = df_term.rename(columns={'OBS_DATE': 'settlement_month'})

# definir target como 1 se POS for SAN (liquidação antecipada) e 0 se SOL (liquidação normal)
# definir settlement_month como a data da última observação (OBS_DATE) para cada contrato, que representa o mês de liquidação (SAN ou SOL)
```

```python
# Juntar settlement_month ao painel completo
temp = BDOSS.merge(
    df_term[['DOSSIER', 'settlement_month']],
    on='DOSSIER',
    how='inner'
)

# Manter apenas observações antes do settlement
temp = temp[temp['OBS_DATE'] < temp['settlement_month']]

# Para cada contrato, ir buscar o último mês antes do settlement
df_prev = (
    temp
    .sort_values(['DOSSIER', 'OBS_DATE'])
    .groupby('DOSSIER', as_index=False)
    .tail(1)
    [['DOSSIER', 'OBS_DATE']]
    .rename(columns={'OBS_DATE': 'feature_month'})
)

# Juntar ao dataset base
o1_base = df_term.merge(df_prev, on='DOSSIER', how='left')

o1_base.head()

# vamos utilizar os dados do mês anterior ao settlement para prever se o contrato terá liquidação antecipada (SAN) ou normal (SOL)
```
